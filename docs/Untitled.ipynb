O que vem dentro do () são os *parâmetros* da função. 
A função `pd.read_csv()` acima tem dois parâmetros: o nome do arquivo e o separador.
O nome do arquivo é um parâmetro obrigatório, porque sem ele o computador não sabe qual arquivo você quer importar. 
Já o parâmetro do separador é opcional mas nós precisamos explicitar, pois o *default* do comando espera um separador com vírgulas, mas o nosso arquivo possui o separador ; 

O comando possui outros vários parâmetros default que estão ocultos, mas que podem ser consultados com o comando `help()`. Isso é especialmente útil quando estamos com problema para importar um arquivo, porque provavelmente algum dos parâmetros default não é compatível com o arquivo que está sendo importado.

Vamos rapidamente só consultar olhar os parâmetros da função `pd.read_csv()` :

```python
help(pd.read_csv)
```
